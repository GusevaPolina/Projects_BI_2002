```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
jellyfish_SRR292678_f_histo = pd.read_table('jellyfish_SRR292678_f.txt', header=None, sep=' ')
jellyfish_SRR292678_r_histo = pd.read_table('jellyfish_SRR292678_r.txt', header=None, sep=' ')
```

```python
jellyfish_SRR292678_f_histo.rename(columns={0: 'count', 1: "raw reads"}, inplace=True)
jellyfish_SRR292678_r_histo.rename(columns={0: 'count', 1: "SRR292678_r"}, inplace=True)
```

```python
j_cor = pd.read_table('/content/hist_j_cor.txt',  header=None, sep=' ')
j_cor.rename(columns={0: 'count', 1: "corrected reads"}, inplace=True)
```

```python
max_f = max(np.array(jellyfish_SRR292678_f_histo["raw reads"])[4:])
# max_r = max(np.array(jellyfish_SRR292678_r_histo["SRR292678_r"])[4:])
max_c = max(np.array(j_cor["corrected reads"])[4:])

max_y_f = jellyfish_SRR292678_f_histo['count'][jellyfish_SRR292678_f_histo["raw reads"] == max_f]
# max_y_r = jellyfish_SRR292678_r_histo['count'][jellyfish_SRR292678_r_histo["SRR292678_r"] == max_r]
max_y_c = j_cor['count'][j_cor["corrected reads"] == max_c]
```

```python
print(max_c, max_f)
```

```python
jellyfish_SRR292678_f_histo['count'][jellyfish_SRR292678_f_histo["raw reads"] == max_f]
```

```python
j_cor['count'][j_cor["corrected reads"] == max_c]
```

```python
from matplotlib.ticker import MultipleLocator
fig, ax = plt.subplots(1,1,figsize=(6,4))

ax.plot(jellyfish_SRR292678_f_histo['count'], 
        jellyfish_SRR292678_f_histo["raw reads"], 
        label='raw reads (SRR292678)', linewidth=4)
ax.plot(j_cor['count'], j_cor["corrected reads"], 
        label='corrected reads (all three libraries)', linewidth=4)
plt.legend(markerscale=6, fontsize=10, loc=3)

r, l = 55, 70
axins = ax.inset_axes([0.4, 0.6, 0.4, 0.3])
axins.scatter(jellyfish_SRR292678_f_histo['count'].loc[r:l], 
              jellyfish_SRR292678_f_histo["raw reads"].loc[r:l], 
              label='raw reads')
axins.scatter(j_cor['count'].loc[r:l], 
              j_cor["corrected reads"].loc[r:l], 
              label='corrected reads')
axins.set_xlim(r, l)
axins.set_ylim(86000, 91000)
axins.set_title('peaks')
ax.indicate_inset_zoom(axins, edgecolor="gray")
axins.tick_params(axis='both', labelsize=11)
axins.xaxis.set_minor_locator(MultipleLocator(2))
axins.grid(which='both')

ax.set_yscale('log')
plt.title('K-mer frequency for raw and processed data')
plt.ylabel('k-mer depth')
plt.xlabel('count for the number of k-mers')
plt.xlim(left=-10, right=300)
plt.ylim(bottom=50)
# plt.grid()
# plt.savefig('sdfs.pdf')
```

As mentioned previously for both SRR292678, 
*   T (total base) is **494 941 140 bp** from FastQC report;
*   M (k-mer peak) is **62** from the k-ner distribution plot;
*   K (k-mer size) is **31 bp** from the settings;
*   L (avg read length) is **90 bp** from FastQC report.

Therefore, N (depth of coverage) is (M x L) / (L - K + 1) = (62 x 90) / (90 - 31 + 1) = 5580 / 60 = **93**

So, genome size is T / N = 494 941 140 / 93 = **5 321 948 bp**
