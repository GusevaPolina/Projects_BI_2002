```python
# let's go and have a look to results
import pandas as pd

with open('result.txt') as result:
  de = pd.read_table(result)

de.head()
```

```python
# label categories of regulation
def label_maker(p,fold):
    part1 = ('Non-s', 'S')[p < 0.05]
    part2 = ('up', 'down')[fold < 0]
    return part1 + 'ignificantly ' + part2 + 'regulated'

de['labels'] = de.apply(lambda x: label_maker(x.padj, x.log2FoldChange), axis=1)
```

```python
dd = de.labels.value_counts().to_dict()
dd
```

```python
lable_list = []

for group in de.labels.unique():
  txt = f'{group}: {dd[group]} ({100 * dd[group] / len(de):.2f}%)'
  print(txt)
  lable_list.append(txt)
```

```python
# plot it and cut the first 50 based on p-values
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(6,4))
label_names = list(de.labels.unique())
for i in range(4):
    lab_inter = de[de['labels'] == label_names[i]]
    plt.scatter(lab_inter['log2FoldChange'], 
                -np.log10(lab_inter['padj']), 
                label=label_names[i], s=2)

plt.legend(lable_list, loc='upper left', markerscale=7)
plt.axvline(x=0, color='gray', linestyle='--')
plt.axhline(-np.log10(0.05), color='gray', linestyle='--')
plt.axhline(-np.log10(de.padj.iloc[50]), color='m', linestyle='--')
plt.text(5.5, -np.log10(0.01), 'p value=0.05')

plt.title('Volcano plot')
plt.ylabel(r'$-log_{10}$ p-value'); plt.xlabel(r'$log_{2}$ fold change')

plt.savefig('volcano.pdf', bbox_inches='tight')
```

```python
# let's do the same but python

de_all = pd.read_table('norm-matrix-deseq2.txt')
de_all.head()
```

```python
from scipy.stats import zscore

gene = de_all.iloc[:, 0]
vals = de_all.iloc[:, 1:].values

# Add noise to avoid zero variance datalines
vals = vals + np.random.normal(0, 0.00001, vals.shape)

# Calculate z-scores
zscore = zscore(vals, axis=1)

# Set gene names as row names
zscore_df = pd.DataFrame(zscore, index=gene, columns=de_all.columns[1:])
```

```python
import seaborn as sns

xticklabels = ['0 minutes\nreplicate 1', '0 minutes\nreplicate 2',
               '30 minutes\nreplicate 1', '30 minutes\nreplicate 2']

cl_map = sns.clustermap(data=zscore_df, cmap='RdYlGn',  # or use z-score=1
                        cbar_kws={'label': 'expression dynamic'},
                        xticklabels=xticklabels, yticklabels=[], 
                        dendrogram_ratio=(0.5, 0.05), figsize=(6, 10), 
                        cbar_pos=(0.02, 0.1, 0.05, 0.28))
cl_map.ax_heatmap.set_ylabel('')
cl_map.fig.suptitle('Clusterisation of gene expression', fontsize=15, y=1.02)

plt.savefig('clustermap_v.pdf', bbox_inches='tight')
```

```python
import seaborn as sns

xticklabels = ['0 minutes\nreplicate 1', '0 minutes\nreplicate 2',
               '30 minutes\nreplicate 1', '30 minutes\nreplicate 2']

cl_map = sns.clustermap(data=zscore_df.T, cmap='RdYlGn', # or use z-score=1
                        yticklabels=xticklabels, xticklabels=[], 
                        dendrogram_ratio=(0.05, 0.5), figsize=(12, 5), 
                        cbar_pos=(0.6, 0.9, 0.3, 0.05), 
                        cbar_kws={"orientation": "horizontal",
                                  'label': 'expression dynamic'})
cl_map.ax_heatmap.set_xlabel('')
cl_map.fig.suptitle('Clustermap of normalized differential expression', 
                    fontsize=15, y=1.02)

plt.savefig('clustermap_h.pdf', bbox_inches='tight')
```

```python
# create a workaround to upload the pathview package into R
with open('installer_pathview.r', 'w') as r_new:
  r_new.write('if (!require("BiocManager", quietly = TRUE))\n')
  r_new.write('install.packages("BiocManager")\n\nBiocManager::install("pathview")')

# run the script with installation
! R -f installer_pathview.r
```

```python
# upload Pathview to R
from rpy2.robjects.packages import importr
import rpy2.robjects as robjects
import rpy2.robjects.packages as rpackages
from rpy2.robjects.vectors import StrVector

import pandas as pd

utils = importr('utils')

# import Pathview
pathview = importr('pathview')

# import top50 genes from DGE
results = pd.read_csv('result_top50.txt', index_col=0)

# create an R object for DGE results
r_results = robjects.r['as.data.frame'](robjects.FloatVector(results.values.flatten()), nrow=len(results), byrow=True)
rownames = robjects.StrVector(results.index)
colnames = robjects.StrVector(results.columns)
r_results.do_slot_assign('dimnames', robjects.ListVector({'Dimnames': [rownames, colnames]}))

# create an R object for gene IDs
gene_ids = robjects.StrVector(results.index)

# run pathview
pathview.pathview(gene_data=r_results, gene_id=gene_ids, pathway_id='hsa00010', species='hsa')
```
